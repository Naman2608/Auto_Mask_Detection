# Mask Detection 


```python
haar_data = cv2.CascadeClassifier('data.xml')
live_img=cv2.VideoCapture(0)
data=[]
font = cv2.FONT_HERSHEY_COMPLEX
while True:
    flag, img = live_img.read()
    if flag:
        faces = haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(22,255,255),4)
            faces=img[y:y+h,x:x+w, : ]
            faces= cv2.resize(faces,(50,50))
            faces = faces.reshape(1,-1)
            #faces= pca.transform(faces)
            pred = svm.predict(faces)
            n = names[int(pred)]
            cv2.putText(img, n, (x,y),font,1,(0,0,255),2)
            print(n)
        cv2.imshow('new tab name',img)
        # 27 - Ascii of Escape
        if cv2.waitKey(2) == 27 or len(data)>=50:
            break
live_img.release()
cv2.destroyAllWindows()
```
